```python
import seaborn as sns
import pandas as pd
```

```python
#Read the dataset 
df = pd.read_csv("penguins_lter.csv")
```

```python
# Show some Headings
df.head()
```

```python
print(df.isnull().sum())
```

```python
df.shape
```

```python
df.drop(columns=["Comments"],inplace = True)
```

```python
df.shape
```

```python
# Replace the null values in sex colomn by mode
df['Sex'] = df['Sex'].fillna(df['Sex'].mode()[0])
```

```python
# There are few rows which are NAN, i take the mean of others and remove it
df['Delta 15 N (o/oo)'] = df['Delta 15 N (o/oo)'].fillna(df['Delta 15 N (o/oo)'].mean())
df['Delta 13 C (o/oo)'] = df['Delta 13 C (o/oo)'].fillna(df['Delta 13 C (o/oo)'].mean())
```

```python
# Removing few rows which are null 
df = df.dropna()
```
