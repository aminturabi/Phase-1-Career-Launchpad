```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("amazon.csv")
```

```python
print(df.shape)
print(df.info())
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True )
```

```python
# Clean Function to remove extra symbols from dataset
def clean_currency(x):
    if isinstance(x, str):
        x = x.replace ('₹', '').replace('%','').replace(',', '').strip()
        try:
            return float(x)
        except ValueError:
            return np.nan
    return x
```

```python
# Applying of cleaning function on dataset
df['discounted_price'] =  df['discounted_price'].apply(clean_currency)
df['actual_price'] =  df['actual_price'].apply(clean_currency)
df['discount_percentage'] =  df['discount_percentage'].apply(clean_currency)
df['rating_count'] =  df['rating_count'].apply(clean_currency)

```

```python
df.head()
```

```python
# how Price, Rating, and Reviews interact
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=[np.number]) # Only numbers for correlation
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation: Do expensive items get better ratings?")
plt.show()
```

```python
# Visual 2: Product Categories
# Which category has the most products?
if 'category' in df.columns:
    plt.figure(figsize=(12, 6))
    # Taking top 10 categories to avoid clutter
    top_categories = df['category'].value_counts().nlargest(10).index
    sns.countplot(y='category', data=df[df['category'].isin(top_categories)], order=top_categories)
    plt.title("Top 10 Product Categories")
    plt.show()
```

```python
# Visual 3: Price Distribution
if 'discounted_price' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df['discounted_price'], kde=True, bins=30)
    plt.title("Distribution of Discounted Prices")
    plt.xlabel("Price")
    plt.show()
```
